```python
import pandas as pd

# Load your dataset into a DataFrame (replace 'your_data.csv' with your dataset file)
df = pd.read_csv("D:/OneDrive/Documents/RESERVED/PERSONAL/COMPUTER TRAININGS/QUANTUM FOLDER/DAF/QA DAF FILES/PYTHON FILES/Unicorn_Companies.csv")

# Display basic information about the dataset
print(df.info())

# Handling Missing Values:
# 1. Drop rows with missing values
df.dropna(inplace=True)

# 2. Fill missing values with a specific value (e.g., fill NaN in a numeric column with the mean)
if 'numerical_cols' in df.columns:
    df['numerical_cols'].fillna(df['numerical_cols'].mean(), inplace=True)

# 3. Fill missing values in a categorical column with the most frequent value
if 'categorical_cols' in df.columns:
    df['categorical_cols'].fillna(df['categorical_cols'].mode()[0], inplace=True)

# Data Integrity:
# 4. Remove duplicate rows
df.drop_duplicates(inplace=True)

# 5. Set appropriate data types for columns
# Replace 'Date_Column' and 'numerical_col' with actual column names
if 'Year Founded' in df.columns:
    df['Year Founded'] = pd.to_datetime(df['Year Founded'])
if 'numerical_cols' in df.columns:
    df['numerical_cols'] = pd.to_numeric(df['numerical_cols'])


# 8. Set the index to a unique identifier column if applicable
# Replace 'Unique_ID_Column' with the actual column name
if 'Unique_ID_Column' in df.columns:
    df.set_index('Unique_ID_Column', inplace=True)

# Final Dataset Info after Data Cleaning:
print(df.info())

# Save the cleaned dataset to a new CSV file
df.to_csv("D:/OneDrive/Documents/RESERVED/PERSONAL/COMPUTER TRAININGS/QUANTUM FOLDER/DAF/QA DAF FILES/PYTHON FILES/Cleaned_Unicorn_Companies.csv", index=False)

# Return the cleaned DataFrame (optional)
df
```

```python
import pandas as pd

# Load your cleaned dataset into a DataFrame
df = pd.read_csv("D:/OneDrive/Documents/RESERVED/PERSONAL/COMPUTER TRAININGS/QUANTUM FOLDER/DAF/QA DAF FILES/PYTHON FILES/Cleaned_Unicorn_Companies.csv")

# Create a cross-tabulation table to analyze the relationship between 'Country' and 'Industry'
cross_tab = pd.crosstab(index=df['Country'], columns=df['Industry'])

# Display the cross-tabulation table
print("Cross-Tabulation of Country and Industry:")
print(cross_tab)

# You can also calculate the percentage of companies in each industry by country
cross_tab_percentage = cross_tab.div(cross_tab.sum(axis=1), axis=0) * 100

# Display the percentage table
print("\nPercentage of Companies in Each Industry by Country:")
print(cross_tab_percentage)

# Visualize the relationship using a heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.heatmap(cross_tab_percentage, annot=True, cmap="YlGnBu")
plt.xlabel('Industry')
plt.ylabel('Country')
plt.title('Relationship between Country and Industry')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your cleaned dataset into a DataFrame
df = pd.read_csv("D:/OneDrive/Documents/RESERVED/PERSONAL/COMPUTER TRAININGS/QUANTUM FOLDER/DAF/QA DAF FILES/PYTHON FILES/Cleaned_Unicorn_Companies.csv")

# Univariate Analysis:
# 1. Distribution of Valuation
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Valuation', bins=20, kde=True)
plt.title('Distribution of Valuation')
plt.xlabel('Valuation')
plt.ylabel('Count')
plt.show()

# 2. Industry Counts
plt.figure(figsize=(10, 6))
df['Industry'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Counts of Companies by Industry')
plt.xlabel('Industry')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()

# Bivariate Analysis:
# 3. Scatter plot between Valuation and Funding
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Valuation', y='Funding', hue='Industry', alpha=0.5)
plt.title('Scatter Plot of Valuation vs. Funding')
plt.xlabel('Valuation')
plt.ylabel('Funding')
plt.show()

# 4. Box plot of Valuation by Continent
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Continent', y='Valuation', palette='Set3')
plt.title('Box Plot of Valuation by Continent')
plt.xlabel('Continent')
plt.ylabel('Valuation')
plt.xticks(rotation=45)
plt.show()

# Multivariate Analysis:
# 5. Correlation Heatmap
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
